```python
# 1. Definir a lista de colunas que não agregam valor preditivo
# O customerID é o principal, mas se houver colunas de índice duplicadas (como 'Unnamed: 0'), adicione-as aqui.
colunas_irrelevantes = ['customerID']

# 2. Remover as colunas do DataFrame original
# Usamos errors='ignore' para evitar que o código trave caso a coluna já tenha sido removida
df_modelo = df.drop(columns=colunas_irrelevantes, errors='ignore')

# 3. Validar a remoção
print(f"✅ Colunas removidas: {colunas_irrelevantes}")
print(f"📊 Novo formato do dataset: {df_modelo.shape} linhas e {df_modelo.shape} colunas.")

# Exibir as colunas restantes que serão usadas no treinamento
print("\nAtributos que serão usados para prever o Churn:")
print(df_modelo.columns.tolist())

# Visualizar o topo do novo DataFrame
display(df_modelo.head())
```

```python
import pandas as pd

# 1. Identificar colunas binárias (apenas 2 opções) e categóricas (3 ou mais)
# Primeiro, vamos mapear a nossa variável alvo (Churn) e outras binárias para 0 e 1
binarias = ['Churn', 'customer.Partner', 'customer.Dependents', 'phone.PhoneService', 'account.PaperlessBilling']

# Mapeamento manual para garantir controle total
for col in binarias:
    if col in df_modelo.columns:
        df_modelo[col] = df_modelo[col].map({'Yes': 1, 'No': 0})

# 2. Aplicar One-Hot Encoding nas demais colunas categóricas
# O Pandas identifica automaticamente as colunas do tipo 'object' ou 'category'
df_final = pd.get_dummies(df_modelo, columns=[
    'customer.gender',
    'phone.MultipleLines',
    'internet.InternetService',
    'internet.OnlineSecurity',
    'internet.OnlineBackup',
    'internet.DeviceProtection',
    'internet.TechSupport',
    'internet.StreamingTV',
    'internet.StreamingMovies',
    'account.Contract',
    'account.PaymentMethod'
], dtype=int) # dtype=int transforma o True/False resultante em 1/0

# 3. Verificação do resultado
print(f"📊 Colunas antes do encoding: {df_modelo.shape}")
print(f"🚀 Colunas após o encoding: {df_final.shape}")

# Visualizar as primeiras linhas do DataFrame pronto para IA
display(df_final.head())
```

```python
from sklearn.model_selection import train_test_split

# 1. Definir os atributos (X) e o alvo (y)
# X contém todas as colunas preditivas (já sem customerID e com Encoding)
X = df_final.drop(columns=['Churn'])

# y é a nossa variável alvo (0 para ficou, 1 para saiu)
y = df_final['Churn']

# 2. Realizar a divisão
# test_size=0.3 define 30% para teste
# random_state=42 garante que os resultados sejam replicáveis
# stratify=y mantém a proporção de 26% de churn em ambos os grupos
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# 3. Verificação das dimensões
print(f"✅ Divisão concluída!")
print(f"Amostras de Treino: {X_train.shape}")
print(f"Amostras de Teste:  {X_test.shape}")

# 4. Validar a proporção de Churn (Stratification)
print("\nProporção de Churn no Treino (%):")
print((y_train.value_counts(normalize=True) * 100).map('{:.2f}%'.format))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# --- PREPARAÇÃO: Modelo 1 (Regressão Logística com Padronização) ---
# Usamos um Pipeline para garantir que a padronização seja aplicada corretamente
model_lr = Pipeline([
    ('scaler', StandardScaler()),
    ('logistic', LogisticRegression(random_state=42))
])

# Treinamento com os dados balanceados (X_res, y_res do passo anterior)
model_lr.fit(X_res, y_res)

# --- PREPARAÇÃO: Modelo 2 (Random Forest - Não exige padronização) ---
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Treinamento (pode usar os dados originais do SMOTE sem scaler)
model_rf.fit(X_res, y_res)

print("✅ Modelos criados e treinados com sucesso!")
print(f"Modelo 1: Regressão Logística (Padronizado)")
print(f"Modelo 2: Random Forest (Original)")
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

def avaliar_modelo(modelo, X_test, y_test, nome_modelo):
    # 1. Predições
    y_pred = modelo.predict(X_test)

    # 2. Métricas principais
    print(f"\n{'='*30}")
    print(f"📊 AVALIAÇÃO: {nome_modelo}")
    print(f"{'='*30}")
    print(classification_report(y_test, y_pred))

    # 3. Matriz de Confusão
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'Matriz de Confusão - {nome_modelo}')
    plt.xlabel('Previsão do Modelo')
    plt.ylabel('Valor Real (Churn)')
    plt.show()

# Rodar avaliação para os dois modelos
avaliar_modelo(model_lr, X_test, y_test, "Regressão Logística")
avaliar_modelo(model_rf, X_test, y_test, "Random Forest")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Extraindo e ACHATANDO os coeficientes (.flatten() transforma 2D em 1D)
coefs = model_lr.named_steps['logistic'].coef_.flatten()
features = X.columns

# 2. Criando o DataFrame (agora as dimensões vão bater: 1D para ambos)
df_coefs = pd.DataFrame({'Feature': features, 'Coefficient': coefs})
df_coefs = df_coefs.sort_values(by='Coefficient', ascending=False)

# 3. Plotando os resultados
plt.figure(figsize=(10, 8))

# Pegando os 10 maiores coeficientes (Vilões do Churn) e os 10 menores (Protetores)
top_viloes = df_coefs.head(10)
top_protetores = df_coefs.tail(10)
df_plot = pd.concat([top_viloes, top_protetores])

colors = ['salmon' if x > 0 else 'skyblue' for x in df_plot['Coefficient']]

plt.barh(df_plot['Feature'], df_plot['Coefficient'], color=colors)
plt.axvline(0, color='black', linestyle='--', linewidth=0.8)
plt.gca().invert_yaxis() # Maior impacto no topo
plt.title('Influência das Variáveis no Churn (Regressão Logística)')
plt.xlabel('Peso do Coeficiente (Positivo = Risco | Negativo = Retenção)')
plt.tight_layout()
plt.show()
```

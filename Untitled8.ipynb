```python
import pandas as pd

# 1. Carregar o ficheiro CSV (ajusta o nome se guardaste com outro nome)
nome_ficheiro = 'telecom_x_limpo.csv'

try:
    df = pd.read_csv(nome_ficheiro)
    print(f"✅ Ficheiro '{nome_ficheiro}' carregado com sucesso!")

    # 2. Eliminar o customerID (caso ainda não o tenhas feito na exportação)
    if 'customerID' in df.columns:
        df = df.drop(columns=['customerID'])
        print("🗑️ Coluna 'customerID' removida para evitar ruído no modelo.")

    # 3. Exibir informações estruturais
    print(f"\nEstrutura Atual: {df.shape} linhas e {df.shape} colunas.")

    # 4. Verificação de tipos de dados (DICA: Devem ser numéricos ou object/category)
    print("\n--- Resumo das Colunas ---")
    print(df.dtypes.value_counts())

    # Visualização das primeiras linhas
    display(df.head())

except FileNotFoundError:
    print(f"❌ Erro: O ficheiro '{nome_ficheiro}' não foi encontrado na pasta atual.")
```

```python
# 1. Lista de colunas a serem removidas
# 'customerID' é a principal, mas se houver outras colunas administrativas, adicione aqui
colunas_para_remover = ['customerID']

# 2. Remover as colunas do DataFrame
df_preditivo = df.drop(columns=colunas_para_remover)

# 3. Verificar o novo formato
print(f"Colunas antes: {df.shape}")
print(f"Colunas atuais: {df_preditivo.shape}")
print("\nColunas mantidas para o modelo:")
print(df_preditivo.columns.tolist())

# 4. Visualizar as primeiras linhas do novo DataFrame
display(df_preditivo.head())
```
